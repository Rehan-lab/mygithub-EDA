## Do people of higher age earn more?


```python
attrition_file.plot(kind = "scatter",        #type of plot
          x = "Age", y = "MonthlyIncome", #variables in x and y axis
         color = "black",          #color of dots
          figsize = (5,5))         #fig size

plt.xlabel("Age",fontsize = 12)
plt.ylabel("Monthly income",fontsize = 12)
plt.title("Scatter Plot",fontsize = 14)

plt.show()
```
